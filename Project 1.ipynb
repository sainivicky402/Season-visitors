```python
columns = ['season','hr','holiday','weekday','weathersit','temp','hum','windspeed','casual','registered','cnt']
weathercount1[columns] = weathercount1[columns].apply(pd.to_numeric)
```

```python
Q1 = weathercount1['registered'].quantile(.25)
Q3 = weathercount1 ['registered'].quantile(.75)
```

```python
def find_outliers(df, columns):
    Q1 = df[columns].quantile(.25)
    Q3 = df[columns].quantile(.75)
    iqr = Q3-Q1
    min_r = Q1 - 1.5*iqr
    max_r = Q3 + 1.5*iqr
    outliers_indices = df.index[(df[columns] < min_r) | (df[columns]> max_r)]
    return outliers_indices
```

```python
find_outliers (weathercount1, 'weathersit')
```

```python
find_outliers (weathercount1, 'windspeed')
```

```python
find_outliers (weathercount1, 'hum')
```

```python
find_outliers (weathercount1, 'casual')
```

```python
find_outliers (weathercount1, 'cnt')
```
